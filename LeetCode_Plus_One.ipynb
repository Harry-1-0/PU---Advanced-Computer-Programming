# 66. Plus One

## Solution 1 - Carry Propagation


```python
# Python solution for leetcode
from typing import List

class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        for i in range(len(digits) - 1, -1, -1):
            if digits[i] < 9:
                digits[i] += 1
                return digits
            digits[i] = 0
        return [1] + digits
```

```python
# Example on test cases
sol = Solution()

# Test case 1
digits_1 = [1,2,3]
solution_1 = sol.plusOne(digits_1)
print(f'Solution for test case 1: {solution_1}')

# Test case 2
digits_2 = [4,3,2,1]
solution_2 = sol.plusOne(digits_2)
print(f'Solution for test case 2: {solution_2}')

# Test case 3
digits_3 = [9]
solution_3 = sol.plusOne(digits_3)
print(f'Solution for test case 3: {solution_3}')
```
