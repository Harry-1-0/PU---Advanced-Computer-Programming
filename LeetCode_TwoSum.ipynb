# 1. Two Sum

## Solution 1

```python
def twoSum_1(nums, target):
    n = len(nums)
    for i in range(n - 1):
        for j in range(i + 1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []
```

```python
nums_1 = [2,7,11,15]
target_1 = 9
print(twoSum_1(nums_1, target_1))
```

## Solution 2

```python
def twoSum_2(nums, target):
    numMap = {}
    n = len(nums)

    for i in range(n):
        complement = target - nums[i]
        if complement in numMap:
            return [numMap[complement], i]
        numMap[nums[i]] = i

    return []
```

```python
nums_2 = [3,2,4]
target_2 = 6
print(twoSum_2(nums_2, target_2))
```
