## 20. Valid Parentheses

### Solution 1: Stack

```python
# Python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        pairs = {')': '(', ']': '[', '}': '{'}
        for char in s:
            if char in pairs.values():
                stack.append(char)
            elif char in pairs:
                if not stack or stack[-1] != pairs[char]:
                    return False
                stack.pop()
        return not stack
```

#### Testcases for Solution 1

```python
sol = Solution()
testcases = ["()", "()[]{}", "(]", "([])", "([)]"]
results = []
for i in range(len(testcases)):
    results.append(sol.isValid(testcases[i]))
for i in range(len(results)):
    print(f'Testcase {i + 1}: Input string: "{testcases[i]}" | Valid Parentheses? - {results[i]}')
```

### Solution 2: Queue (Doubly Ended Queue)

```python
# Python
from collections import deque

class Solution:
    def isValid(self, s: str) -> bool:
        queue = deque()
        pairs = {')': '(', ']': '[', '}': '{'}
        for char in s:
            if char in pairs.values():
                queue.append(char)
            elif char in pairs:
                if not queue or queue[-1] != pairs[char]:
                    return False
                queue.pop()

        return not queue
```

#### Testcases for Solution 2

```python
sol = Solution()
testcases = ["()", "()[]{}", "(]", "([])", "([)]"]
results = []
for i in range(len(testcases)):
    results.append(sol.isValid(testcases[i]))
for i in range(len(results)):
    print(f'Testcase {i + 1}: Input string: "{testcases[i]}" | Valid Parentheses? - {results[i]}')
```
