## 704. Binary Search

### Solution - Python

```python
class Solution(object):
    def search(self, nums, target):
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return -1
```

#### Test Cases

```python
sol = Solution()
nums_1 = [-1,0,3,5,9,12]
target_1 = 9
print(f'Testcase 1: Input array: {nums_1} | Target number: {target_1} | Index: {sol.search(nums_1, target_1)}')
nums_2 = [-1,0,3,5,9,12]
target_2 = 2
print(f'Testcase 2: Input array: {nums_2} | Target number: {target_2} | Index: {sol.search(nums_2, target_2)}')
```
